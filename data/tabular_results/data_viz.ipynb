## Data Visualization for Filter Performance:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```


 - #### LMS

```python
# LMS
lms_noises = [
    ("air_conditioner",    "./air_conditioner/LMS_results.csv",       "./air_conditioner/LMS_ALE_results.csv"),
    ("babble",             "./babble/LMS_results.csv",               "./babble/LMS_ALE_results.csv"),
    ("cafe",               "./cafe/LMS_results.csv",                 "./cafe/LMS_ALE_results.csv"),
    ("munching",           "./munching/LMS_results.csv",             "./munching/LMS_ALE_results.csv"),
    ("typing",             "./typing/LMS_results.csv",               "./typing/LMS_ALE_results.csv"),
    ("washer_dryer",       "./washer_dryer/LMS_results.csv",         "./washer_dryer/LMS_ALE_results.csv"),
]

lms_ref_vals = []
ale_vals = []

for label, path_lms, path_lms_ale in lms_noises:
    df_lms = pd.read_csv(path_lms)
    df_lms_ale = pd.read_csv(path_lms_ale)
    lms_ref_vals.append(df_lms[f"LMS Mean Delta SNR: {label} noise"].iloc[-1])
    ale_vals.append(df_lms_ale[f"LMS Mean Delta SNR: {label} noise"].iloc[-1])

x = np.arange(len(lms_noises))
width = 0.35

fig, ax = plt.subplots(figsize=(10,6))
non_ale = ax.bar(x - width/2, lms_ref_vals, width, label="Noise Reference", color="darkviolet", alpha=0.88)
ale = ax.bar(x + width/2, ale_vals, width, label="ALE", color="crimson", alpha=0.88)

# setting the layout proper
ax.set_xticks(x)
ax.set_xticklabels([l for l, *_ in lms_noises], rotation=0,)
ax.set_xlabel("Noise types")
ax.set_ylabel("Delta SNR (dB)")
ax.set_ylim(-1, 20)
ax.set_title("Mean Delta SNR by noise type: LMS")
ax.legend()
ax.grid(axis="y", linestyle=":", alpha=0.75)

fig.tight_layout()
plt.show()
```
